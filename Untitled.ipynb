```python
unique_emotion = df["emotion"].unique()
unique_emotion
```

```python
emotion_numbers = {
    'joy': 0,
    'fear': 1,
    'anger': 2,
    'sadness': 3,
    'surprise': 4,
    'love': 5
}
df['emotion'] = df['emotion'].map(emotion_numbers)
```

```python
df
```

```python
df["text"]=df["text"].apply(lambda x: x.lower())
```

```python
df
```

```python
import string
def remove_punc(txt):
    return txt.translate(str.maketrans('','',string.punctuation))
df["text"]=df["text"].apply(remove_punc)
```

```python
df
```

```python
def remove_number(txt):
    new=""
    for i in txt:
        if not i.isdigit():
            new = new +i
    return new
df["text"]=df["text"].apply(remove_number)
```

```python
df
```

```python
def remove_emoji(txt):
    new=""
    for i in txt:
        if i.isascii():
            new = new +i
    return new
df["text"]=df["text"].apply(remove_emoji)
```

```python
df
```

```python
import nltk 
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
nltk.download("punkt",)
nltk.download("stopwords", download_dir='c:/NLP/nltk_data')
```

```python
nltk.download("punkt")
```

```python
stopword=set(stopwords.words("english"))
```

```python
stopword
```

```python
def remove(txt):
    words = txt.split()
    cleaned=[]
    for i in words:
        if not i in stopword:
            cleaned.append(i)

    return ' '.join(cleaned)
```

```python
df["text"]=df["text"].apply(remove)
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df["text"],df["emotion"],test_size=0.20)
```

```python
a=1
emotion_number={}
for i in unique_emotion:
    emotion_number[i]=a
    a+=1
df["emotion"]=df["emotion"].map(emotion_number)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer,CountVectorizer
```

```python
bow_vectorizer=CountVectorizer()
x_train_bow=bow_vectorizer.fit_transform(x_train)
x_test_bow=bow_vectorizer.transform(x_test)
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
```

```python
nb_model=MultinomialNB()
```

```python
nb_model.fit(x_train_bow,y_train)
```

```python
y_pred=nb_model.predict(x_test_bow)
score=accuracy_score(y_test,y_pred)
score
```

```python
tfidf_vectorizer=TfidfVectorizer()
x_train_tfidf=tfidf_vectorizer.fit_transform(x_train)
x_test_tfidf=tfidf_vectorizer.transform(x_test)
```

```python
nb_model_tfidf=MultinomialNB()
nb_model_tfidf.fit(x_train_tfidf,y_train)
```

```python
y_pred_tfidf=nb_model_tfidf.predict(x_test_tfidf)
acc=accuracy_score(y_test,y_pred_tfidf)
acc
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train_tfidf,y_train)
```

```python
y_pred_2=model.predict(x_test_tfidf)
acc2=accuracy_score(y_test,y_pred_2)
acc2
```
